# MENGGUNAKAN INPUT USER

```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Algoritma Rekursif untuk SOPRTS
def recursive_schedule(locations, route=None):
    if route is None:
        route = []
    if not locations:
        return route
    next_location = min(locations)
    route.append(next_location)
    locations.remove(next_location)
    return recursive_schedule(locations, route)

# Algoritma Iteratif untuk SOPRTS
def iterative_schedule(locations):
    sorted_route = []
    while locations:
        next_location = min(locations)
        sorted_route.append(next_location)
        locations.remove(next_location)
    return sorted_route

# Fungsi untuk mengukur waktu eksekusi
def measure_time(algorithm, locations):
    start_time = time.time()
    algorithm(locations[:])  # Gunakan salinan untuk menjaga daftar asli
    return time.time() - start_time

# Data untuk grafik dan tabel
locations_count = []
recursive_times = []
iterative_times = []

# Fungsi untuk mencetak tabel hasil
def print_table():
    table = PrettyTable()
    table.field_names = ["Jumlah Lokasi", "Waktu Rekursif (detik)", "Waktu Iteratif (detik)"]
    for i in range(len(locations_count)):
        table.add_row([locations_count[i], f"{recursive_times[i]:.6f}", f"{iterative_times[i]:.6f}"])
    print("\nTabel Perbandingan Kinerja:")
    print(table)

# Grafik kinerja
def plot_performance():
    plt.figure(figsize=(10, 6))
    plt.plot(locations_count, recursive_times, label='Rekursif', marker='o', color='blue')
    plt.plot(locations_count, iterative_times, label='Iteratif', marker='s', color='orange')
    plt.title('Perbandingan Kinerja: Rekursif vs Iteratif')
    plt.xlabel('Jumlah Lokasi')
    plt.ylabel('Waktu Eksekusi (detik)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Program utama
while True:
    try:
        n = int(input("Masukkan jumlah lokasi (atau ketik -1 untuk keluar): "))
        if n == -1:
            print("Program selesai. Grafik terakhir akan ditampilkan.")
            plot_performance()
            print_table()
            break
        if n <= 0:
            print("Masukkan jumlah lokasi yang valid!")
            continue

        # Membuat daftar lokasi dari 1 hingga n
        locations = list(range(1, n + 1))
        locations_count.append(n)

        # Mengukur waktu untuk algoritma rekursif
        recursive_time = measure_time(recursive_schedule, locations)
        recursive_times.append(recursive_time)

        # Mengukur waktu untuk algoritma iteratif
        iterative_time = measure_time(iterative_schedule, locations)
        iterative_times.append(iterative_time)

        # Tampilkan hasil sementara
        print(f"\nJumlah Lokasi: {n}")
        print(f"Waktu Rekursif: {recursive_time:.6f} detik")
        print(f"Waktu Iteratif: {iterative_time:.6f} detik")
        print_table()  # Perbarui tabel setiap iterasi

    except ValueError:
        print("Masukkan jumlah lokasi yang valid!")
```

# TIDAK MENGGUNAKAN INPUT USER

```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Algoritma Rekursif untuk penjadwalan truk sampah
def recursive_schedule(locations, route=None):
    if route is None:
        route = []
    if not locations:
        return route
    next_location = min(locations)
    route.append(next_location)
    locations.remove(next_location)
    return recursive_schedule(locations, route)

# Algoritma Iteratif yang lebih efisien untuk penjadwalan truk sampah
def iterative_schedule(locations):
    route = []
    while locations:
        next_location = min(locations)
        route.append(next_location)
        locations.remove(next_location)
    return route

# Grafik untuk menyimpan data
locations_count = []
recursive_times = []
iterative_times = []

# Fungsi untuk memperbarui grafik
def update_graph():
    plt.figure(figsize=(8, 6))
    plt.plot(locations_count, recursive_times, label='Recursive', marker='o', linestyle='-')
    plt.plot(locations_count, iterative_times, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative')
    plt.xlabel('Number of Locations')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table():
    table = PrettyTable()
    table.field_names = ["Number of Locations", "Recursive Time (s)", "Iterative Time (s)"]
    for i in range(len(locations_count)):
        table.add_row([locations_count[i], recursive_times[i], iterative_times[i]])
    print(table)

# Program utama tanpa input
def main():
    for n in range(1, 1001, 100):
        locations = list(range(1, n + 1))
        locations_count.append(n)

        # Ukur waktu eksekusi algoritma Rekursif
        start_time = time.time()
        recursive_schedule(locations[:])
        recursive_times.append(time.time() - start_time)

        # Ukur waktu eksekusi algoritma Iteratif
        start_time = time.time()
        iterative_schedule(locations[:])
        iterative_times.append(time.time() - start_time)

    # Cetak tabel waktu eksekusi
    print_execution_table()

    # Perbarui grafik
    update_graph()

if __name__ == "__main__":
    main()
```
